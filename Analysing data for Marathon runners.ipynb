```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")

df.head(10)
```

```python
df.dtypes
```

```python
#Data Cleaning
#50km and 50mi need to be combined

df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
#filter for 2020
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
#filter by country



df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA' )]
df2.head(10)
```

```python
#clean up age and small details
df2['athlete_age'] = 2024 - df2['Athlete year of birth']
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
#get rid of some columns: CLub, country,year of birth, Age category

df2 = df2.drop(['Year of event', 'Athlete club' , 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
#null cleanup
df2.isna().sum()
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
#duplicate check
df2[df2.duplicated() == True]
```

```python
#reset index
df2.reset_index(drop = True)
```

```python
#fix data types
df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
df2.dtypes
```

```python
#fix column names to make for ease of use

df2 = df2.rename(columns = {'Event dates':'race_day',
                            'Event name':  'race_name',                 
                            'Event distance/length': 'race_length',        
                            'Event number of finishers':  'race_number_of_finishers',   
                            'Athlete performance': 'athlete_performance',          
                            'Athlete gender': 'gender',                
                            'Athlete average speed': 'average_speed',        
                            'Athlete ID': 'id'                                      
                            })
```

```python
df2.head(10)
```

```python
# Lets do some filtering
df2[df2['race_name'] == 'West Seattle Beach Run - Winter Edition (USA)']
```

```python
#Graph

sns.histplot(df2['race_length'])
```

```python
sns.lmplot(data = df2, x = 'athlete_age', y = 'average_speed', hue = 'gender')
```

```python
#difference in speed for 50km, 50mi for male to female

df2.groupby(['race_length', 'gender'])['average_speed'].mean()
```

```python
#what age groups are the best in 50mi race

df2.query('race_length == "50mi"').groupby('athlete_age')['average_speed'].agg(['mean', 'count']).sort_values('mean',ascending = False).query('count>19')
```

```python
#what age groups are the worst in 50mi race

df2.query('race_length == "50mi"').groupby('athlete_age')['average_speed'].agg(['mean', 'count']).sort_values('mean',ascending = True).query('count>9')
```

```python
#are runners slower in the summer than winter?

df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x : 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df2.groupby('race_season')['average_speed'].agg(['mean', 'count']).sort_values('mean' ,ascending = False)
```

```python
#Query with 50km only
df2.query('race_length == "50km"').groupby('race_season')['average_speed'].agg(['mean', 'count']).sort_values('mean' ,ascending = False)
```

```python
#Query with 50mi only
df2.query('race_length == "50mi"').groupby('race_season')['average_speed'].agg(['mean', 'count']).sort_values('mean' ,ascending = False)
```
